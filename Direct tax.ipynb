## Distribution of Direct Tax Collection by Govt.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
Direct_taxes = pd.read_csv('data.csv')
```

```python
Direct_taxes
```

```python
Direct_taxes.info()
```

```python
Direct_taxes['Year'] = Direct_taxes['Year'].str.replace('*', '')
```

```python
Direct_taxes.columns
```

```python
columns_to_convert = ['Corporate Tax', 'Income Tax', 'Other tax', 'Total']
```

```python
Direct_taxes[columns_to_convert] = Direct_taxes[columns_to_convert].replace(',', '', regex = True)
```

```python
 Direct_taxes['Total']  = Direct_taxes['Total'].replace('#', '', regex = True)
```

```python
Direct_taxes[columns_to_convert] = Direct_taxes[columns_to_convert].apply(pd.to_numeric)
```

```python
Direct_taxes
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.set_style('whitegrid')
plt.plot(Direct_taxes['Year'],Direct_taxes['Corporate Tax'], marker = 'o', label = 'Corporate Tax')
plt.plot(Direct_taxes['Year'],Direct_taxes['Income Tax'], marker = 'o', label = 'Income Tax')
plt.plot(Direct_taxes['Year'],Direct_taxes['Total'], marker = 'o', label = 'Total Tax')
plt.legend()
plt.ylabel('Values in Crores')
plt.title('Net Direct Tax Collection (2013-2023)')
# to not have values in the form of scientific notation on the y axis
plt.ticklabel_format(style = 'plain', axis = 'y')
# Remove the lines along the x-axis and y-axis
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.show();
```
